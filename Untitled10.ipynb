```python
def make_heatmap(numeric_df, outpath):
    corr = numeric_df.corr()
    mask = np.triu(np.ones_like(corr, dtype=bool))
    plt.figure(figsize=(max(10, corr.shape[0]*0.35), max(8, corr.shape[0]*0.25)))
    sns.set(style="white")
    cmap = sns.diverging_palette(10, 220, as_cmap=True)
    sns.heatmap(corr, mask=mask, cmap=cmap, vmax=1.0, vmin=-1.0, center=0,
                square=False, linewidths=.5, annot=True, fmt=".2f", cbar_kws={"shrink": .6})
    plt.title("Correlation matrix (numeric features)")
    plt.tight_layout()
    # Save THEN show
    plt.savefig(outpath, dpi=200, bbox_inches="tight")
    plt.show()
    plt.close('all')
```

```python
# Cell 2 - check for danceability and points-like column names
cols_lower = {c.lower(): c for c in df.columns}
dance_col = None
points_col = None

# common possible names
for name in ("danceability", "danceability_score", "dance"):
    if name in cols_lower:
        dance_col = cols_lower[name]
        break
if dance_col is None:
    # fallback: try to find a column with 'dance' in its name
    for k,v in cols_lower.items():
        if "dance" in k:
            dance_col = v
            break

for name in ("points", "Points", "score", "total_points", "points_awarded"):
    if name.lower() in cols_lower:
        points_col = cols_lower[name.lower()]
        break
if points_col is None:
    for k,v in cols_lower.items():
        if "point" in k or "score" in k:
            points_col = v
            break

if dance_col is None or points_col is None:
    raise SystemExit(f"Could not find both danceability and points columns. Found danceability={dance_col}, points={points_col}")

print("Using columns:", dance_col, "and", points_col)

# keep only rows with non-missing values for the two columns
df_plot = df[[dance_col, points_col]].copy()
before = len(df_plot)
df_plot = df_plot.dropna(subset=[dance_col, points_col])
after = len(df_plot)
print(f"Dropped {before-after} rows with missing danceability or points. Remaining rows: {after}")
```

```python
# Cell 3 - compute Pearson r and make plot + text pane saved as a single PNG
import scipy.stats as stats
sns.set(style="whitegrid")

# Convert to numeric if needed
df_plot[dance_col] = pd.to_numeric(df_plot[dance_col], errors='coerce')
df_plot[points_col] = pd.to_numeric(df_plot[points_col], errors='coerce')
df_plot = df_plot.dropna(subset=[dance_col, points_col])

# compute Pearson correlation
r, p = stats.pearsonr(df_plot[dance_col], df_plot[points_col])
n = len(df_plot)
r_text = f"Pearson r = {r:.3f}\np-value = {p:.3e}\nn = {n}"

# Create a figure with two panels: left scatter+regression, right text
fig, axes = plt.subplots(ncols=2, figsize=(12,5), gridspec_kw={'width_ratios': [2,1]})

# Left: scatter + regression line
ax = axes[0]
sns.scatterplot(x=dance_col, y=points_col, data=df_plot, ax=ax, s=40, alpha=0.7, edgecolor='k')
# fit linear regression line
slope, intercept, r_val, p_val_lin, std_err = stats.linregress(df_plot[dance_col], df_plot[points_col])
x_vals = np.linspace(df_plot[dance_col].min(), df_plot[dance_col].max(), 100)
ax.plot(x_vals, intercept + slope*x_vals, color='red', lw=2, label=f"fit: y={slope:.2f}x+{intercept:.1f}")
ax.set_title("Danceability vs Points")
ax.legend(loc='best')
ax.grid(True)

# annotate correlation on the scatter plot
ax.text(0.02, 0.95, r_text, transform=ax.transAxes, fontsize=10, verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

# Right: hypothesis / interpretation text
ax2 = axes[1]
ax2.axis('off')
# Formulate a brief hypothesis based on computed r
if abs(r) < 0.1:
    relation = "no meaningful linear relationship"
    hypothesis = ("Hypothesis: Danceability does not have a meaningful linear relationship\n"
                  "with points received. Other factors (performance, staging, geopolitics)\n"
                  "likely explain variation in points.")
elif r > 0:
    if r < 0.3:
        strength = "a weak positive"
    elif r < 0.6:
        strength = "a moderate positive"
    else:
        strength = "a strong positive"
    relation = f"{strength} relationship (positive)"
    hypothesis = (f"Hypothesis: Songs with higher danceability tend to receive higher points\n"
                  f"({relation}). Possible reason: more danceable songs are more engaging and\n"
                  f"appeal to televoters and juries, but effect size is limited.")
else:
    if r > -0.3:
        strength = "a weak negative"
    elif r > -0.6:
        strength = "a moderate negative"
    else:
        strength = "a strong negative"
    relation = f"{strength} relationship (negative)"
    hypothesis = (f"Hypothesis: Songs with higher danceability tend to receive lower points\n"
                  f"({relation}). Possible reason: more danceable songs might be perceived as\n"
                  f"less musically complex or less emotionally resonant by juries.")

# Compose text for display
text = f"Selected features: {dance_col} (x) and {points_col} (y)\n\n{r_text}\n\nInterpretation:\n{hypothesis}"
ax2.text(0, 0.98, text, fontsize=11, va='top')

plt.tight_layout()
out_png = "danceability_vs_points_with_hypothesis.png"
plt.savefig(out_png, dpi=200, bbox_inches='tight')
plt.show()

print("Saved combined image to", out_png)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error

# Define X and y
# Assuming 'Points' is the target variable (y)
y = df_work['Points']
# X will be all numeric columns in df_work, excluding 'Points' and 'Unnamed: 0'
# which typically serves as an index and not a feature.
# Also exclude any identified categorical columns that might have slipped through
# or are not intended as features for a numeric model.
features_to_exclude = ['Points', 'Unnamed: 0'] + cat_cols # cat_cols was defined earlier
X = df_work.drop(columns=[col for col in features_to_exclude if col in df_work.columns], errors='ignore').select_dtypes(include=np.number)

# Create a dummy args object for test_size and random_state
class Args:
    def __init__(self):
        self.test_size = 0.2
        self.random_state = 42
args = Args()

# train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=args.test_size, random_state=args.random_state)
print(f"Train size: {len(y_train)}, Test size: {len(y_test)}")

# Train Gradient Boosting Regressor
model = GradientBoostingRegressor(random_state=args.random_state, n_estimators=200, learning_rate=0.05, max_depth=3)
model.fit(X_train, y_train)

# Predict & evaluate MAE
y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
print(f"\nMean Absolute Error (MAE) on test set: {mae:.2f} points")
```
